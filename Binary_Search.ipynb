```python
def locate_card(cards,query):
  lo,hi=0, len(cards)-1

  #getting the middle card
  while lo<=hi:
    mid=(lo+hi) //2
    mid_number=cards[mid]
    print("hi:",hi,"lo:",lo)

    result=test_location(cards,query,mid)

    if result == 'found':
      return mid
    elif result =='left':
      return mid-1
    elif result == 'right':
      return mid+1


  return -1


```
