```python
class treeNode():
  def __init__(self,key):
    self.key=key
    self.left=None
    self.right=None
```

```python
def parse_tuple(data):
  print(data)
  if isinstance( data,tuple) and len(data)==3:#if data is type of tuple and___
    node=treeNode(data[1])#setting the root as 2,,
    node.left=parse_tuple(data[0])#(1,3,None),,
    node.right=parse_tuple(data[2])#((None,3,4),5,(6,7,8))),,
  elif data == None:
    node = None #here((1,3,None)
  else:
    node=treeNode(data)#simply create a node as 1  ||| this is the termination condition for the tree
  return node
```

```python
def disply_keys(node,space='\t',level=0):


       #if node is empty
       if node is None:
        print(space*level +'0')
        return

        #if node is the leaf node
       if node.left is None and node.right is None:
         print(space*level + str(node.key))
         return

       #if the node has children(use recursion)
       disply_keys(node.right,space,level+1)
       print(space*level + str(node.key))
       disply_keys(node.left,space,level+1)



```

```python
tree = parse_tuple(((1,3,None), 2, ((None, 3, 4), 5, (6, 7, 8))))
```

```python
disply_keys(tree)
```

### **Implementing Inorder Traversal**

```python
def traverse_inorder(node):
  if node is None:#this is the terminating condition
    return []

  return(traverse_inorder(node.left)+
         [node.key]+
         traverse_inorder(node.right))
```

```python
traverse_inorder(tree)
```

### **Implementing Preorder Traversal**

```python
def traverse_preorder(node):
  if node is None:
    return []

  return([node.key]+
         traverse_preorder(node.left)+
         traverse_preorder(node.right))
```

```python
traverse_preorder(tree)
```

### **Implementing postorder Traversal**

```python
def traverse_postorder(node):
  if node is None:
    return []

  return(traverse_postorder(node.left)+
         traverse_postorder(node.right)+
                            [node.key])
```

```python
traverse_postorder(tree)
```

### **Calculating the height of the binary tree**

```python
def height(node):
  if node is None:
    return 0

  return 1+max(height(node.left),height(node.right))
```

```python
height(tree)
```

### **Count the number of nodes in a tree**

```python
def count(node):
  if node is None:
    return 0

  return 1 + count(node.left)+count(node.right)
```

```python
count(tree)
```
